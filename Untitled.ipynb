```python
import pandas as pd
import numpy as np
import mne

import matplotlib.pyplot as plt
```

```python
dataset_path ='Tekin_MI_LR_Training 1.csv'
raw_csv = pd.read_csv(dataset_path,sep='\t',header=None)
```

```python
raw_csv
```

```python
raw_eeg_csv = raw_csv.iloc[1:,3:5]
raw_eeg_csv
```

```python
raw_eeg_csv.rename(columns={3:'C3',4:'C4'})
```

```python
C3_values = raw_eeg_csv.iloc[1000:,0:1].values.ravel()
C4_values = raw_eeg_csv.iloc[1000:,1:2].values.ravel()
t = np.linspace(start=0,stop=len(C3_values),num=len(C4_values))
```

```python
fig,ax = plt.subplots(3,1)

ax[1].axis('off')

ax[0].plot(t,C3_values)
ax[2].plot(t,C4_values)
```

```python
plt.plot(t,C3_values)
```

```python
plt.plot(t,C4_values)
```

```python
info_st = mne.create_info(ch_names = ['C3','C4'],sfreq=255)
raw = mne.io.RawArray((C3_values,C4_values),info=info_st)
```
